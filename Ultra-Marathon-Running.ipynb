## Cleaning Data

#### Only USA Races 

```python
df['Event distance/length'].unique()
```

#### 50mi or 50km

```python
df[df['Event distance/length'].isin(['50km','50mi'])]
```

#### for the year 2020


```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) ]
```

```python
df[df['Event name'].str.contains('USA')]
```

```python
# Combining all conditions

df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.contains('USA')) ]
```

```python
df2.shape
```

```python
df2.head()
```

```python
# remove (USA) from Event name
df2['Event name'].str.split('(')
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
# clean Athelete age category
df2['Athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
df2.head()
```

```python
df2.shape
```

```python
# remove h from athlete performance
df2['Athlete performance'] = df2['Athlete performance'].str.split('h').str.get(0)
```

```python
df2.head()
```

```python
# drop columns which are not required
df2 = df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis = 1)
df2.head()
```

```python
df2.shape
```

```python
# cleaning null values

df2.isna().sum()
```

```python
df2[df2['Athlete_age'].isna() == 1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
df2.isnull().sum()
```

```python
# check duplicates
df2[df2.duplicated() == True]
```

```python
# reset index
df2.reset_index(inplace = True)
```

```python
df2
```

```python
# fix types
df2.dtypes
```

```python
df2['Athlete_age'] = df2['Athlete_age'].astype(int)
```

```python
df2.dtypes
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
# rename columns
df2.rename(columns = {'Year of event' : 'year',
                       'Event dates' : 'race_day',
                     'Event name' : 'race_name',
                     'Event distance/length' : 'race_length',
                     'Event number of finishers' : 'race_number_of_finishers',
                     'Athlete performance' : 'athlete_performance',
                     'Athlete gender' : 'athlete_gender',
                     'Athlete average speed' : 'athlete_average_speed',
                     'Athlete ID' : 'athlete_id',
                     'Athlete_age' : 'athlete_age'}, inplace = True)
```

```python
df2.columns
```

```python
# reorder columns
df3 = df2[['race_day','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed']]
```

```python
df3
```

# Questions to find out from the data

#### Difference in speed for 50km, 50mi male to female

```python
df3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

#### what age groups are the best in the 50mi race ( 30+ races min) and top 10

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>29').head(10)
```

#### what age groups are the best in the 50mi race ( 30+ races min) and bottom 10

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = True).query('count>29').head(10)
```

#### Athlete slower in Summer than Winter?

```python
# spring 3-5
# summer 6-8
# fall 9-11
# winter 12 - 2
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head()
```

```python
# df3.loc[(df3['race_month'] > 2) & (df3['race_month'] < 6), 'Season'] = 'spring'
# df3.loc[(df3['race_month'] > 5) & (df3['race_month'] < 9), 'Season'] = 'summer'
# df3.loc[(df3['race_month'] > 8) & (df3['race_month'] < 12), 'Season'] = 'fall'
# df3.loc[(df3['race_month'] > 11) | (df3['race_month'] < 3), 'Season'] = 'spring'

df3['Season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df3.head()
```

```python
df3.groupby('Season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending=False)
```
